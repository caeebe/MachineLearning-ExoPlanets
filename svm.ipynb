```python
import pandas as pd
```

# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("cumulative.csv")
df = df.drop(columns=["rowid", "kepid", "kepoi_name", "kepler_name", "koi_pdisposition", "koi_score", "koi_tce_delivname"])
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head()
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
df.shape
```

```python

y = df["koi_disposition"]
X = df.drop("koi_disposition", axis=1)

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=50)
```

```python
df.groupby(["koi_disposition"]).count()
```

```python
X_train.head()
```

# Pre-processing

Scale the data using the MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler
X_scaler = MinMaxScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

# Train the Support Vector Machine

```python
from sklearn.svm import SVC 
model = SVC(kernel='rbf', gamma = "scale", C=10000)
model.fit(X_train_scaled, y_train)
modelL=SVC(kernel='linear', C = 10000)
modelL.fit(X_train_scaled, y_train)
```

```python
print(f"RBF Training Data Score: {model.score(X_train_scaled, y_train)}")
print(f"RBF Testing Data Score: {model.score(X_test_scaled, y_test)}")
print(f"Linear Training Data Score: {modelL.score(X_train_scaled, y_train)}")
print(f"Linear Testing Data Score: {modelL.score(X_test_scaled, y_test)}")
```

# Hyperparameter Tuning

Use `GridSearchCV` to tune the `C` and `gamma` parameters
- C : float, optional (default=1.0) Penalty parameter C of the error term
- kernel : string, optional (default=â€™rbfâ€™) Specifies the kernel type to be used in the algorithm. It must be one of â€˜linearâ€™, â€˜polyâ€™, â€˜rbfâ€™, â€˜sigmoidâ€™, â€˜precomputedâ€™ or a callable
- gamma : float, optional (default=â€™autoâ€™) Kernel coefficient for â€˜rbfâ€™, â€˜polyâ€™ and â€˜sigmoidâ€™. Default is â€˜autoâ€™ which uses 1 / n_features, if gamma='scale' is passed then it uses 1 / (n_features * X.var()) as value of gamma


```python
# Create the GridSearchCV model
from sklearn.model_selection import GridSearchCV
param_grid = [{'kernel': ['rbf'], 'C': [1, 10, 100, 1000, 5000], 'gamma': [0.001, 0.01, 0.1, "scale"]},
             {'kernel': ['linear'], 'C': [1, 10, 100, 1000, 5000]},
             {'kernel': ['poly'], 'C':[1, 10, 100, 1000, 5000], 'gamma': [.001, .01, .1, 'scale']}]

grid = GridSearchCV(SVC(), param_grid, verbose=3)
```

```python
# Train the model with GridSearch
grid.fit(X_train_scaled, y_train)
```

```python
print(grid.best_params_)
print(grid.best_score_)
```

```python
# Loop through different k values to see which has the highest accuracy
# Note: We only use odd numbers because we don't want any ties
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
train_scores = []
test_scores = []
for k in range(1, 20, 2):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train_scaled, y_train)
    train_score = knn.score(X_train_scaled, y_train)
    test_score = knn.score(X_test_scaled, y_test)
    train_scores.append(train_score)
    test_scores.append(test_score)
    print(f"k: {k}, Train/Test Score: {train_score:.3f}/{test_score:.3f}")
    
    
plt.plot(range(1, 20, 2), train_scores, marker='o')
plt.plot(range(1, 20, 2), test_scores, marker="x")
plt.xlabel("k neighbors")
plt.ylabel("Testing accuracy Score")
#plt.show()
```

```python
print(f"Best Parameters Data Score: {grid.score(X_train_scaled, y_train)}")
print(f"Best Parameters Testing Score: {grid.score(X_test_scaled, y_test)}")
```
